```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
Avviation = pd.read_csv('AviationData.csv', encoding='latin-1');
Avviation.head(7).T
```

```python
mean_mode_values = {
    'Total.Uninjured': Avviation['Total.Uninjured'].mean(),
    'Total.Minor.Injuries': Avviation['Total.Minor.Injuries'].mean(),
    'Total.Serious.Injuries': Avviation['Total.Serious.Injuries'].mean(),
    'Total.Fatal.Injuries': Avviation['Total.Fatal.Injuries'].mean(),
    'Purpose.of.flight':Avviation['Purpose.of.flight'].mode()[0],
    'Air.carrier':Avviation['Air.carrier'].mode()[0],
    'Aircraft.Category':Avviation['Aircraft.Category'].mode()[0],
    'FAR.Description':Avviation['FAR.Description'].mode()[0],
    'Schedule':Avviation['Schedule'].mode()[0],
    'Air.carrier':Avviation['Air.carrier'].mode()[0],
    'Broad.phase.of.flight':Avviation['Broad.phase.of.flight'].mode()[0],
    'Engine.Type':Avviation['Engine.Type'].mode()[0],
    'Number.of.Engines':int(Avviation['Number.of.Engines'].mean()),
    'Weather.Condition':Avviation['Weather.Condition'].mode()[0],
    'Airport.Code':Avviation['Airport.Code'].mode()[0],
    'Airport.Name':Avviation['Airport.Name'].mode()[0],
    'Aircraft.damage':Avviation['Aircraft.damage'].mode()[0],
    'Injury.Severity':Avviation['Injury.Severity'].mode()[0],
    'Make':Avviation['Make'].mode()[0],
    'Amateur.Built':Avviation['Amateur.Built'].mode()[0],
    'Model':Avviation['Model'].mode()[0],
    'Report.Status':Avviation['Report.Status'].mode()[0],
    'Registration.Number':Avviation['Registration.Number'].mode()[0],
    'Country':Avviation['Country'].mode()[0],
    'Location':Avviation['Location'].mode()[0]
}
Avviation.fillna(value=mean_mode_values, inplace=True)
# cleaning data and replaing the data with the relavent method for object's and numerics 
# replacing mean for numerical null values and mode for object types
```
